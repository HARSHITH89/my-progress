```python
from IPython.display import HTML

html_content = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Tap Progress Tracker</title>
    <style>
        /* CSS: This styles our webpage to make it look nice */
        body {
            font-family: Arial, sans-serif;
            display: flex;
            justify-content: center;
            align-items: center;
            height: 100vh;
            background-color: #f4f4f9;
            margin: 0;
        }
        .container {
            background-color: white;
            padding: 30px;
            border-radius: 15px;
            box-shadow: 0 4px 8px rgba(0,0,0,0.1);
            text-align: center;
            width: 300px;
        }
        .input-group {
            margin-bottom: 20px;
        }
        input[type="number"] {
            width: 80px;
            padding: 5px;
            font-size: 16px;
        }
        .stats {
            font-size: 24px;
            margin: 20px 0;
            font-weight: bold;
        }
        #tapBtn {
            background-color: #28a745;
            color: white;
            border: none;
            padding: 15px 30px;
            font-size: 20px;
            border-radius: 5px;
            cursor: pointer;
            width: 100%;
            margin-bottom: 10px;
        }
        #tapBtn:active {
            background-color: #218838; /* Makes the button look "pressed" */
        }
        #resetBtn {
            background-color: #dc3545;
            color: white;
            border: none;
            padding: 10px 20px;
            border-radius: 5px;
            cursor: pointer;
        }
    </style>
</head>
<body>

    <div class="container">
        <h2>Tap Tracker</h2>

        <div class="input-group">
            <label for="targetInput">Set Target: </label>
            <input type="number" id="targetInput" value="100" min="1">
        </div>

        <div class="stats">
            <div>Taps: <span id="tapCount">0</span></div>
            <div style="color: #007bff; margin-top: 10px;">Progress: <span id="percentage">0</span>%</div>
            <div style="color: #ffc107; margin-top: 10px;">Taps per Hour: <span id="tapsPerHourDisplay">0.0</span></div>
            <div style="color: #17a2b8; margin-top: 10px;">Total Hours: <span id="totalHoursDisplay">0.00</span></div>
        </div>

        <button id="tapBtn">TAP ME!</button>
        <button id="resetBtn">Reset</button>
    </div>

    <script>
        // 1. Grab the elements from the screen
        let count = 0;
        let startTime = null; // New variable for tracking start time
        const targetInput = document.getElementById('targetInput');
        const tapCountDisplay = document.getElementById('tapCount');
        const percentageDisplay = document.getElementById('percentage');
        const tapsPerHourDisplay = document.getElementById('tapsPerHourDisplay'); // New element for Taps per Hour
        const totalHoursDisplay = document.getElementById('totalHoursDisplay'); // New element for Total Hours
        const tapBtn = document.getElementById('tapBtn');
        const resetBtn = document.getElementById('resetBtn');

        // 2. Function to update the math and screen
        function updateScreen() {
            // Get the target number from the input box
            let target = parseInt(targetInput.value);

            // Prevent dividing by zero or negative numbers if the user deletes the target
            if (isNaN(target) || target <= 0) {
                target = 1;
            }

            // Calculate percentage: (Taps / Target) * 100
            let percent = (count / target) * 100;

            // Update the text on the screen
            tapCountDisplay.textContent = count;

            // toFixed(1) keeps it to 1 decimal place (e.g., 5.5% instead of 5.55555%)
            percentageDisplay.textContent = percent.toFixed(1);

            // Logic for Taps per Hour and Total Hours
            let elapsedTimeInSeconds = 0;
            if (startTime === null && count > 0) {
                startTime = Date.now();
            } else if (count === 0) {
                startTime = null;
            }

            let tapsPerHour = 0.0;
            let totalHours = 0.0;

            if (startTime !== null) {
                elapsedTimeInSeconds = (Date.now() - startTime) / 1000;
                if (elapsedTimeInSeconds > 0) {
                    tapsPerHour = (count / elapsedTimeInSeconds) * 3600;
                    totalHours = elapsedTimeInSeconds / 3600;
                }
            }
            tapsPerHourDisplay.textContent = tapsPerHour.toFixed(1);
            totalHoursDisplay.textContent = totalHours.toFixed(2);
        }

        // 3. What happens when you click "TAP ME!"
        tapBtn.addEventListener('click', () => {
            count++; // Increase count by 1
            updateScreen(); // Run the math
        });

        // 4. What happens when you click "Reset"
        resetBtn.addEventListener('click', () => {
            count = 0; // Reset count to 0
            startTime = null; // Reset start time
            updateScreen(); // Run the math
        });

        // 5. Update math instantly if the user changes the target number
        targetInput.addEventListener('input', () => {
            updateScreen();
        });

        // Initial screen update to set values
        updateScreen();
    </script>
</body>
</html>
"""
HTML(html_content)
```
